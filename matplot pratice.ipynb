```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
##to show plots in notebooks
```

```python
import numpy as np
x=np.linspace(0,5,11)
y=x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y,'r-')
#plt.show()#only use when not using in notebook
plt.xlabel('x label')
plt.ylabel('y label')
plt.title('title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
## object orinted method
fig=plt.figure()

axes=fig.add_axes([1,1,.8,1])

axes.plot(x,y)
axes.set_xlabel('x lable')
axes.set_xlabel('y label')
axes.set_title('title')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,.1,.8,.9])
axes2=fig.add_axes([0.2,0.3,0.2,0.5])

axes1.plot(x,y)
axes2.plot(y,x)
```

```python
#subplots with object methods
fig,axes=plt.subplots(nrows=1,ncols=2)
plt.tight_layout()
axes[0].plot(x,y)
axes[0].set_title('x plot')
axes[1].plot(y,x)
axes[1].set_title('y plot')
```

# figure size and dpi

```python
fig=plt.figure(figsize=(8,2),dpi=100)

ax=fig.add_axes([0,0,1,1])

ax.plot(x,y)
```

```python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X squared')
ax.plot(x,x**3,label='X cube')

ax.legend(loc=(0.1,0.1))
```

plot appearance

```python
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',linewidth=3,alpha=10,linestyle='-',marker='o',markersize=30,
       markerfacecolor='yellow',markeredgewidth=0.9,markeredgecolor='green')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
