```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

# Distribution plots

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.distplot(tips['tip'],kde=False,bins=40)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
    sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
sns.rugplot(tips['total_bill'])
```

# categorical plots

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)#x is catergorical but y will be numeric
```

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',split=True)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips)#not use this forlarge datasets
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='violin')
```

# matrix plots 

```python
flights=sns.load_dataset('flights')
```

```python
flights.head()
```

```python
tc=tips.corr()
```

```python
sns.heatmap(tc,annot=True,cmap='coolwarm')
```

```python
fp=flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(fp,cmap='coolwarm',linewidths=3,linecolor='white')
```

```python
sns.clustermap(fp,cmap='coolwarm',standard_scale=1)#to get similar betwen data
```

# Grids

```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
g=sns.PairGrid(iris)
#g.map(plt.scatter)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
g=sns.FacetGrid(data=tips,col='time',row='smoker')
g.map(sns.distplot,'total_bill')
g.map(plt.scatter,'total_bill','tip')
```

# regression plots

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],
          scatter_kws={'s':100}
          )
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time',aspect=0.6,size=6)
```
